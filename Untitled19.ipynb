```python
import zipfile
with zipfile.ZipFile('/content/dataset.zip', 'r') as zf:
  zf.extractall('/content/dataset')
```

```python
import os
dataset_images = os.path.join('/content/dataset', 'images')
os.listdir(dataset_images)
```

```python
import tensorflow as tf
import keras.layers as layers
from keras.preprocessing.image import ImageDataGenerator
```

```python
#view some of images of every cricketer
import cv2
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(12, 12))
for i in range(len(os.listdir(dataset_images))):
  ax = fig.add_subplot(5, 3, i+1)
  cricketer_dir = os.path.join(dataset_images, os.listdir(dataset_images)[i])
  image_path = os.listdir(cricketer_dir)[0]
  img = plt.imread(os.path.join(cricketer_dir, image_path))
  img_rez = cv2.resize(img, (720, 540))
  plt.axis('off')
  ax.set_xlabel(os.listdir(dataset_images)[i])
  plt.imshow(img_rez)
```

```python
#preprocess and make a training and validation data
class_names = os.listdir(dataset_images)
image_size = (128, 128)
batch_size = 32
seed = 0
datagen = ImageDataGenerator(rescale=1/255, 
                             validation_split=0.2)
train_datagen = datagen.flow_from_directory(dataset_images,
                      
                                            subset='training',
                                            target_size=image_size,
                                            seed=seed,
                                            batch_size=batch_size
                                            
                                            )
val_datagen = datagen.flow_from_directory(dataset_images,
                                            subset='validation',
                                            target_size=image_size,
                                            seed=seed,
                                            batch_size=batch_size)
```

```python
#model architecture
model = tf.keras.Sequential()

model.add(layers.Conv2D(32, kernel_size=3, padding='same', activation='relu', input_shape=(128, 128, 3)))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))
model.add(layers.BatchNormalization())


model.add(layers.Conv2D(64, kernel_size=3, padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))
model.add(layers.BatchNormalization())


model.add(layers.Conv2D(128, kernel_size=3, padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))
model.add(layers.BatchNormalization())


model.add(layers.Conv2D(256, kernel_size=3, padding='same', activation='relu'))
model.add(layers.MaxPooling2D(pool_size=(2, 2), padding='same'))
model.add(layers.BatchNormalization())

model.add(layers.Flatten())
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(512, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dropout(0.3))
model.add(layers.Dense(15, activation='softmax'))

#compiling the model
model.compile(optimizer='adam', metrics='accuracy', loss='categorical_crossentropy')
model.summary()
```

```python
#earlyStopping callback
early_stopping = tf.keras.callbacks.EarlyStopping(
    monitor="val_loss",
    patience=5,
    # mode="auto",
    restore_best_weights=True,
)
history = model.fit(train_datagen,
                    validation_data=val_datagen,
                    callbacks=[early_stopping],
                    epochs=10,
                    batch_size=32)
```

```python
import pandas as pd
df = pd.DataFrame(history.history)
df.plot()
```

```python
#using inbuilt mobilenet model for object detection
mobilenet = tf.keras.applications.MobileNetV2(weights="imagenet",include_top=False, input_shape=(128, 128, 3))
mobilenet.summary()
```

```python
#we won't train mobilenet network from scratch
#and just add our ANN on top of it
for layer in mobilenet.layers:
    layer.trainable = False
```

```python
#using keras functional API
flattened_output = tf.keras.layers.Flatten()(mobilenet.output)
output1 = tf.keras.layers.Dense(1024, activation="relu")(flattened_output)
output2 = tf.keras.layers.Dense(512, activation="relu")(output1)
output3 = tf.keras.layers.Dense(512, activation="relu")(output2)
dropout1 = tf.keras.layers.Dropout(0.3)(output3)
output4 = tf.keras.layers.Dense(256, activation="relu")(dropout1)
dropout2 = tf.keras.layers.Dropout(0.3)(output4)
output5 = tf.keras.layers.Dense(128, activation="relu")(dropout2)
dropout3 = tf.keras.layers.Dropout(0.3)(output5)
output6 = tf.keras.layers.Dense(15, activation="softmax")(dropout3)
model = tf.keras.Model(inputs=mobilenet.input, outputs=output6)
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=["accuracy"])

history = model.fit(train_datagen,
                    validation_data=val_datagen,
                    # callbacks=[early_stopping],
                    epochs=20,
                    batch_size=16)
```

```python
#clearly our model is overfitting now try to use another model to reduce overfitting
flattened_output = tf.keras.layers.Flatten()(mobilenet.output)
output1 = tf.keras.layers.Dense(1024, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.001))(flattened_output)
output2 = tf.keras.layers.Dense(512, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.001))(output1)
dropout0 = tf.keras.layers.Dropout(0.3)(output2)
output3 = tf.keras.layers.Dense(512, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.001))(dropout0)
dropout1 = tf.keras.layers.Dropout(0.35)(output3)
output4 = tf.keras.layers.Dense(256, activation="relu", kernel_regularizer=tf.keras.regularizers.l2(0.001))(dropout1)
dropout2 = tf.keras.layers.Dropout(0.4)(output4)
output6 = tf.keras.layers.Dense(15, activation="softmax")(dropout2)
model2 = tf.keras.Model(inputs=mobilenet.input, outputs=output6)
model2.summary()

```

```python
model2.compile(loss='categorical_crossentropy', 
              optimizer='adam',
              metrics=["accuracy"])

from keras.callbacks import ModelCheckpoint

model_checkpoint = ModelCheckpoint("model_weights.h5", monitor='val_accuracy', verbose=1, save_best_only=True, mode='auto')

history = model2.fit(train_datagen,
                    validation_data=val_datagen,
                    callbacks=[model_checkpoint],
                    epochs=10,
                    batch_size=16)
```
